```python
import pandas as pd 
```

## get Data

```python
pairs = pd.read_csv("data/day4.in", delimiter='[,-]'
, names=["elf1_start", "elf1_end","elf2_start","elf2_end"]
)
pairs.head()
```

## Question 1

```python
pairs['2_in_1'] = (pairs['elf2_start'] >= pairs['elf1_start']) & (pairs['elf2_end'] <= pairs['elf1_end'])
pairs['1_in_2'] = (pairs['elf1_start'] >= pairs['elf2_start']) & (pairs['elf1_end'] <= pairs['elf2_end'])
pairs['one_in']     = (pairs['1_in_2'] | pairs['2_in_1'])
```

```python
pairs.head(20)

print(pairs['one_in'].sum())
```

## Question 2

```python
pairs['overlap']  = ~(pairs['elf1_end'] < pairs['elf2_start']) & ~(pairs['elf2_end'] < pairs['elf1_start'])
pairs.head(20)
```

```python
print(pairs['overlap'].sum())
```
